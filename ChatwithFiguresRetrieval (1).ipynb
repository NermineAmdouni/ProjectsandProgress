```python
from langchain_community.document_loaders import PyPDFLoader
import fitz  # PyMuPDF
import re
import os
import base64
from PIL import Image
import io
pdf_path = "/content/drive/MyDrive/PMBOK.pdf"
class BookFigureExtractor:
    def __init__(self, pdf_path):
        self.pdf_path = pdf_path
        self.figures = []

    def extract_figures_with_text(self):
        """Extract figures using text pattern matching"""
        doc = fitz.open(self.pdf_path)
        figures = []

        for page_num in range(len(doc)):
            page = doc[page_num]
            text = page.get_text()

            # Pattern to match figure references
            # Adjust these patterns based on your book's format
            patterns = [
                r'Figure\s+(\d+[-.]?\d*)\s*[:.]\s*([^\n]+)',  # Standard figure caption
                r'Fig\.\s*(\d+[-.]?\d*)\s*[:.]\s*([^\n]+)',   # Abbreviated figure reference
                r'FIGURE\s+(\d+[-.]?\d*)\s*[:.]\s*([^\n]+)'   # Uppercase variant
            ]

            for pattern in patterns:
                matches = re.finditer(pattern, text)
                for match in matches:
                    figure_num = match.group(1)
                    caption = match.group(2).strip()

                    figure_data = {
                        'figure_number': figure_num,
                        'caption': caption,
                        'page_number': page_num + 1,
                        'text_context': self._get_surrounding_text(text, match.start(), 200)
                    }
                    figures.append(figure_data)

        self.figures = figures
        return figures

    def extract_figures_with_images(self):
        """Extract actual figure images from PDF"""
        doc = fitz.open(self.pdf_path)
        figures_with_images = []

        for page_num in range(len(doc)):
            page = doc[page_num]
            image_list = page.get_images(full=True)

            # Extract text to match images with captions
            text = page.get_text()

            for img_index, img in enumerate(image_list):
                xref = img[0]
                base_image = doc.extract_image(xref)
                image_bytes = base_image["image"]

                # Convert image bytes to base64 for storage/display
                image_b64 = base64.b64encode(image_bytes).decode()

                # Try to match image with nearby caption
                caption = self._find_nearest_caption(text, img_index)

                figure_data = {
                    'page_number': page_num + 1,
                    'image_base64': image_b64,
                    'image_format': base_image["ext"],
                    'caption': caption,
                    'position_on_page': img_index
                }
                figures_with_images.append(figure_data)

        return figures_with_images

    def _get_surrounding_text(self, text, position, context_length):
        """Get text surrounding a figure reference"""
        start = max(0, position - context_length)
        end = min(len(text), position + context_length)
        return text[start:end].strip()

    def _find_nearest_caption(self, text, img_position):
        """Find the nearest figure caption to an image"""
        caption_pattern = r'Figure\s+(\d+[-.]?\d*)\s*[:.]\s*([^\n]+)'
        matches = list(re.finditer(caption_pattern, text))

        if matches:
            # Simple heuristic: return the caption closest to the image position
            # You might want to improve this based on your book's specific layout
            return matches[min(img_position, len(matches)-1)].group(2).strip()
        return "Caption not found"

    def save_figures_to_neo4j(self, neo4j_graph, include_images=False):
        """Save extracted figures to Neo4j"""
        for figure in (self.figures if not include_images
                      else self.extract_figures_with_images()):
            # Create Cypher query
            query = """
            CREATE (f:Figure {
                number: $figure_number,
                caption: $caption,
                page: $page_number,
                context: $context
            })
            """

            params = {
                'figure_number': figure.get('figure_number', 'unknown'),
                'caption': figure.get('caption', ''),
                'page_number': figure.get('page_number'),
                'context': figure.get('text_context', '')
            }

            if include_images and 'image_base64' in figure:
                query = query.replace('})', ', image: $image})')
                params['image'] = figure['image_base64']

            neo4j_graph.query(query, params)

    def export_figures_to_markdown(self, output_dir):
        """Export figures to markdown format"""
        os.makedirs(output_dir, exist_ok=True)

        with open(os.path.join(output_dir, 'figures.md'), 'w') as f:
            for figure in self.figures:
                f.write(f"## Figure {figure['figure_number']}\n\n")
                f.write(f"**Caption:** {figure['caption']}\n\n")
                f.write(f"**Page:** {figure['page_number']}\n\n")
                f.write(f"**Context:**\n{figure['text_context']}\n\n")
                f.write("---\n\n")

# Example usage:
def process_book_figures(pdf_path, neo4j_graph=None):
    extractor = BookFigureExtractor(pdf_path)

    # Extract figures with text
    figures = extractor.extract_figures_with_text()

    # Optionally extract images
    figures_with_images = extractor.extract_figures_with_images()



    # Export to markdown
    extractor.export_figures_to_markdown('book_figures')

    return figures, figures_with_images

```

```python
def find_relevant_figures(figures, question):
    """Find figures relevant to the question"""
    if not figures:
        return []

    # Convert question to lowercase for matching
    question_lower = question.lower()
    question_words = set(question_lower.split())

    relevant_figures = []
    for figure in figures:
        # Check if any question words appear in caption or context
        caption_lower = figure['caption'].lower()
        context_lower = figure['text_context'].lower()

        # Calculate relevance score based on word matches
        score = sum(1 for word in question_words
                   if word in caption_lower or word in context_lower)

        if score > 0:
            relevant_figures.append({
                'figure': figure,
                'score': score
            })

    # Sort by relevance score and take top 3
    relevant_figures.sort(key=lambda x: x['score'], reverse=True)
    return [item['figure'] for item in relevant_figures[:3]]
```

```python
import fitz  # PyMuPDF
import cv2
import numpy as np
from PIL import Image
import io
import re
import os
import pytesseract  # For OCR text extraction
from google.colab import files
import shutil

# Load figure titles from 'figures_titles.txt'
titles_file = "/content/drive/MyDrive/List_Figures_Tables.txt"  # Adjust the path if necessary
with open(titles_file, 'r') as file:
    figure_titles = [line.strip() for line in file.readlines()]

# Load the PDF document
pdf_document = "/content/drive/MyDrive/PMBOK.pdf"  # Replace with your file path
pdf = fitz.open(pdf_document)

# Create a directory for saving the extracted figures
output_dir = "/content/extracted_figures/"
os.makedirs(output_dir, exist_ok=True)

# Function to extract OCR text from the image
def extract_text_from_image(image):
    # Ensure the image is not empty
    if image.size == 0:
        return ""

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Use Tesseract OCR to extract text
    text = pytesseract.image_to_string(gray)
    return text

# To keep track of used titles and prevent duplicates
used_titles = set()
title_index = 0  # Initialize an index for matching figures with titles

for page_number in range(32, pdf.page_count + 1):  # Start from page 32 as requested
    page = pdf.load_page(page_number - 1)  # PyMuPDF uses 0-based indexing
    pix = page.get_pixmap(dpi=300)  # Render at 300 DPI for high resolution
    img_bytes = pix.tobytes(output="png")

    # Convert image bytes to a NumPy array for OpenCV processing
    image = Image.open(io.BytesIO(img_bytes))
    image_np = np.array(image)

    # Convert to grayscale for contour detection
    gray = cv2.cvtColor(image_np, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    edges = cv2.Canny(blurred, 50, 150)

    # Find contours in the image
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Filter contours to find the figure's black border
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)

        # Filter based on the size and aspect ratio (adjust based on your figures' characteristics)
        if w > 500 and h > 300:  # Adjust these values based on your figure size
            figure = image_np[y:y+h, x:x+w]

            # Convert the cropped figure back to an image
            figure_image = Image.fromarray(figure)

            # Extract OCR text from the figure's surrounding area (above or below the figure)
            surrounding_text = ""

            # Check if the region above the figure has valid pixels
            if y - 100 >= 0:  # Ensure we're not going out of bounds
                surrounding_text += extract_text_from_image(image_np[y-100:y, x:x+w])  # Text above the figure

            # Check if the region below the figure has valid pixels
            if y + h + 100 <= image_np.shape[0]:  # Ensure we're not going out of bounds
                surrounding_text += extract_text_from_image(image_np[y+h:y+h+100, x:x+w])  # Text below the figure

            # Try to match the extracted surrounding text with a title in the list
            for title in figure_titles:
                if title.lower() in surrounding_text.lower() and title not in used_titles:
                    # Title found, save the figure with the title in the filename
                    figure_filename = f"{title}.png"

                    # Sanitize filename to remove characters that are not allowed in filenames
                    figure_filename = re.sub(r'[\\/*?:"<>|]', "_", figure_filename)  # Replace invalid characters with underscores

                    # Save the figure as a PNG file
                    figure_image.save(os.path.join(output_dir, figure_filename))
                    print(f"Extracted figure from page {page_number} saved as {figure_filename}.")

                    # Mark this title as used
                    used_titles.add(title)
                    title_index += 1
                    break  # Stop once the first matching title is found
    else:
        print(f"No figure matching the criteria was found on page {page_number}.")

# After extracting all figures, provide the user with a way to download the images
if os.listdir(output_dir):
    # Create a zip file of the extracted figures
    shutil.make_archive("/content/extracted_figures", 'zip', output_dir)
    print("Figures extracted and saved successfully.")

    # Download the zip file containing the figures
    files.download("/content/extracted_figures.zip")
else:
    print("No figures found matching the criteria.")
```

```python
import os
import base64
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Path to the figures folder
figures_folder = "/content/extracted_figures"

# Path to the text file containing figure titles
figures_txt_path = "/content/drive/MyDrive/List_Figures_Tables.txt"  # Update this to your file path

# Load sentence transformer model for contextual understanding
model = SentenceTransformer('all-MiniLM-L6-v2')  # Small and efficient model for embeddings

# Query for which you want to find relevant figures
query = "What is the relation between Value delivery and Project?"

# Step 1: Load figure titles from the text file
def load_figures_from_txt(file_path):
    with open(file_path, "r") as f:
        figures_list = f.readlines()
    figures_list = [figure.strip() for figure in figures_list]  # Clean up whitespace
    return figures_list

figures_list = load_figures_from_txt(figures_txt_path)

# Step 2: Compute embeddings for both the query and figure titles
def compute_embeddings(text_list):
    return model.encode(text_list)

# Encode query and figure titles
query_embedding = model.encode([query])[0]  # Encode query
figure_embeddings = compute_embeddings(figures_list)  # Encode all figure titles

# Step 3: Compute cosine similarity between query and figure titles
def compute_similarity(query_embedding, figure_embeddings):
    return cosine_similarity([query_embedding], figure_embeddings)[0]

similarity_scores = compute_similarity(query_embedding, figure_embeddings)

# Step 4: Rank figures based on similarity scores
def rank_figures_by_similarity(figures_list, similarity_scores, top_k=2):
    ranked_results = []
    for i, score in enumerate(similarity_scores):
        ranked_results.append((figures_list[i], score))

    # Sort by similarity score in descending order and take only the top K results
    ranked_results.sort(key=lambda x: x[1], reverse=True)
    return ranked_results[:top_k]

# Rank figures and take top 2 most relevant
ranked_figures = rank_figures_by_similarity(figures_list, similarity_scores, top_k=2)

# Step 5: Avoid processing duplicate figures
def remove_duplicates(results):
    seen = set()
    unique_results = []
    for result in results:
        if result[0] not in seen:
            unique_results.append(result)
            seen.add(result[0])
    return unique_results

ranked_figures = remove_duplicates(ranked_figures)

# Step 6: Find and display relevant figures with image previews
def display_relevant_figures(figures_folder, ranked_figures):
    all_figure_files = os.listdir(figures_folder)

    for figure_title, score in ranked_figures:
        print(f"Matched figure: {figure_title} (score: {score:.2f})")
        matched_files = [file for file in all_figure_files if figure_title in file]

        if matched_files:
            figure_path = os.path.join(figures_folder, matched_files[0])
            with open(figure_path, "rb") as image_file:
                image_base64 = base64.b64encode(image_file.read()).decode('utf-8')
            print(f"Image Base64 (preview): {image_base64[:100]}...")  # Display first 100 characters of base64
            print("-" * 80)
        else:
            print(f"No matching image file found for {figure_title}")

# Execute the steps
if ranked_figures:
    display_relevant_figures(figures_folder, ranked_figures)
else:
    print("No relevant figures matched the query.")
```
